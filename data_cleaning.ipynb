```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('election_results.xlsx')
df
```

```python
to_drop = ['Incumbent', 'Biden% 2020', 'Trump% 2020', 'Clinton% 2016', 'Trump% 2016']
df.drop(columns=to_drop, inplace=True, axis=1)
df
```

```python
df.head()
```

```python
df['Winner 2020'] = df.apply(lambda row: 0 if row['Biden 2020'] > row['Trump 2020'] else 1, axis=1)
df['Winner 2016'] = df.apply(lambda row: 0 if row['Clinton 2016'] > row['Trump 2016'] else 1, axis=1)

to_drop = ['Party', 'Biden 2020', 'Trump 2020', 'Clinton 2016', 'Trump 2016', 'Total 2016']
df.drop(columns=to_drop, inplace=True, axis=1)
```
